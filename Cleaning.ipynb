```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
print(customer_df.head())
print(customer_df.columns)

```

```python
if 'customer_id' in customer_df.columns:
    customer_df = customer_df.drop(columns=['customer_id'])
else:
    print("'customer_id' column not found in DataFrame")
```

```python
numerical_df = customer_df.select_dtypes(include=['number'])
categorical_df = customer_df.select_dtypes(exclude=['number'])
```

```python
for col in categorical_df.columns:
    print(f'Value counts for {col}:\n{customer_df[col].value_counts()}\n')
```

```python
target_column = 'Customer Lifetime Value'  # Replace with your actual target column name

selected_columns = categorical_df.columns[:3]  # Select a few categorical columns for plotting

for col in selected_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=col, y=customer_df[target_column], data=customer_df)
    plt.title(f'{col} vs {target_column}')
    plt.xticks(rotation=45)
    plt.show()
```
